````python
import os
import sys
import pandas as pd
from prep_collection import PrepCollection as prep
````

````python
path_input_relative = "/Datasets/Gender Bias/106-workplace-sexism/ISEP Sexist Data labeling.xlsx"
path_output_relative = "/Preprocessed_Datasets/106-workplacesexism.csv"
````

````python
wdr_path = os.path.dirname(os.path.dirname(os.getcwd()))
df = pd.read_excel(os.path.join(wdr_path + path_input_relative))

df['id'] = pd.DataFrame(range(len(df))) + 1
df['text'] = df['Sentences']
df['text'] = df['text'].apply(str).apply(prep.prepare_text)

df['label'] = df['Label']

df = df.reindex(columns=['id','text','label'])
df
````

````python
df.to_csv(wdr_path + path_output_relative)
````
