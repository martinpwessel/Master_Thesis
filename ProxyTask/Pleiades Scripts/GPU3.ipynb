```python
#ELECTRA
from transformers import ElectraForSequenceClassification, ElectraTokenizer
electra_tokenizer = ElectraTokenizer.from_pretrained('google/electra-base-discriminator')
electra_model = ElectraForSequenceClassification.from_pretrained('google/electra-base-discriminator')
proxy_electra = ProxyTraining(electra_tokenizer, 4)
proxy_electra.proxy_train_main(electra_model, 32, gpu_no=3, model_name= "Electra")
```

```python
# DialoGPT
torch.cuda.empty_cache()
from transformers import AutoTokenizer, AutoModelForSequenceClassification
dialogpt_tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-small")
dialogpt_tokenizer.pad_token = dialogpt_tokenizer.eos_token
dialogpt_model = AutoModelForSequenceClassification.from_pretrained('microsoft/DialoGPT-small', num_labels=2)
dialogpt_model.config.pad_token_id = dialogpt_tokenizer.pad_token_id
proxy_dialogpt = ProxyTraining(dialogpt_tokenizer, 4)
proxy_dialogpt.proxy_train_main(dialogpt_model, 8, gpu_no= 3, model_name= "DialoGPT")
```
